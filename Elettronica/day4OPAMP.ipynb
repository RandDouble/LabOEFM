```python
import numpy as np
import matplotlib.pyplot as  plt
import pandas as pd


R1 = 1.01e3 #Ohm
R2 = 9.94e3 #Ohm
V_in = 1 #V
```

```python
V_out = -R2/R1*V_in
V_out
```

## Elaborazione Dati

Raccogliamo dati, grafico in scala log.

```python
%matplotlib inline
df = pd.read_csv('day4OPAMP.csv')
df.head()
plt.style.use('seaborn')
plt.plot(df['FREQ'], 20*np.log10(df['V_OUT']/df['V_IN']), '-o')
plt.xscale('log')
plt.xlabel('frequency [Hz]')
plt.ylabel('Gain [dB]')
plt.title('Gain Bode Diagram')
plt.grid(b=True, which='minor', color='w', linewidth=0.5)
```

```python
%matplotlib inline
df = pd.read_csv('day4NotInvOPAMP.csv')
data_1=df[df['R_2_2']<13e3]
data_2=df[df['R_2_2']>13e3]

plt.plot(data_1['FREQ'], 20*np.log10(data_1['V_OUT']/data_1['V_IN']), '-o', label='Resistenze Bilanciate')
plt.plot(data_2['FREQ'], 20*np.log10(data_2['V_OUT']/data_2['V_IN']), '-o', label='Resistenze Non Bilanciate')
plt.xscale('log')
plt.xlabel('frequency [Hz]')
plt.ylabel('Gain [dB]')
plt.title('Gain Bode Diagram')
plt.legend()
plt.grid(b=True, which='minor', color='w', linewidth=0.5)
```
