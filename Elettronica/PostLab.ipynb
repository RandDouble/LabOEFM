```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Circuito RLC

Per circuito risonante con RLC Serie necessario calcolo parametro:
$$
    \zeta = \frac{R}{2}\sqrt{\frac{C}{L}}
$$

Infatti se $\zeta<1$ si avrà una risonanza, in alternativa si avrà un filtro passa alto.

```python
def calc_zeta(R,L,C):
    return R/2*np.sqrt(C/L)

R = 150 # ohm
L = 123e-6
C = np.logspace(-9, -6, 200)
C_1=13.8e-9
C_2 = 680e-12
print(f'taglio: {1/(np.sqrt(L*C_1)*2*np.pi):.2g} Hz')
print(f'risonanza: {1/(np.sqrt(L*C_2)*2*np.pi):.2g} Hz')

plt.plot(C, calc_zeta(R,L,C))
plt.xscale('log')
plt.axhline(1)
plt.xlabel('Capacità [F]')
plt.ylabel('Zeta')
plt.grid(True, 'both')
```

# Circuito Colpitts

Bisogna verificare la condizione di Barkhausen:

$$
    G\ge\frac{1}{\Beta}=\frac{C_1}{C_2}\\
    G = \frac{R_2}{R_1}\\
    \frac{R_2}{R_1}\ge\frac{C_1}{C_2}
$$

```python
Rin = 50 #Ohm
Rout= 500 #Ohm
C_1 = np.logspace(-12, -6, 200)
C_2 =  100e-9

plt.semilogx(C_1, C_1/C_2)
plt.axhline(Rout/Rin)
plt.grid(True, 'both')
```
