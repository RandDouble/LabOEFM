# Verifica Legge di Ohm
## Preparazione dei Dati
Importo i moduli necessari per l'analisi e i data contenuti in un file csv

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

%matplotlib inline

data = pd.read_csv("data.csv")
```

##  Calcolo del valore di resistenza dalla pendenza della retta
In output si ha anche il valore della pendenza della retta

```python
%matplotlib inline
x = data["Tensione [V]"].values.reshape(-1, 1)
y = data["Corrente [A]"].values.reshape(-1,1)

regr=LinearRegression()
regr.fit(x,y)
plt.scatter(x,y)
plt.plot(x, regr.predict(x), '--r')
plt.xlabel('Tensione [V]')
plt.ylabel('Corrente [A]')
plt.title("Verifica della Legge di Ohm")
print('Valore Resistenza : {} Ohm'.format(1/regr.coef_))
```

## Sezione per Calcolare Partitore Resistenze

In questa sezione vado a calcolare le tensioni generate nel partitore, sono richieste la $V_in$, $R_1$ e $R_2$

```python
def Partitore(Vin : float, R1 : float, R2 : float ):
    V1 = Vin *(R1/(R1+R2))
    V2 = Vin *(R2/(R1+R2))
    return V1, V2

R1 = 2200 # Ohm
R2 = 1000 # Ohm
Vin = 150e-3 # V

Partitore(Vin, R1, R2)
```
